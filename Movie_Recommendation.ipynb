```python
import numpy as np
import pandas as pd
```

```python
movie = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie.head(1).T
```

```python
credits.head(1).T
```

```python
movies=movie.merge(credits , on='title')
```

```python
movies.head(2).T
```

```python
movies.shape
```

```python
movies = movies[['genres','keywords','overview','title','movie_id','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def covert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(covert)
```

```python
movies['keywords'] = movies['keywords'].apply(covert)
```

```python
def covert3(obj):
    l = []
    count=0
    for i in ast.literal_eval(obj):
        if count !=3:
            l.append(i['name'])
            count +=1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(covert3)
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies = movies[['movie_id','title','tags']]
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:" ".join(x)) # to convert list into string
```

```python
new_movies.head()
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:x.lower())
                         
```

```python
new_movies.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split(): # to convert string into list
        y.append(ps.stem(i))
    return " ".join(y)  # convrt list into string
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vactors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vactors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vactors)
```

```python
def recommended(movie):
    movie_index = new_movies[new_movies['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommended("Batman Begins")
```
